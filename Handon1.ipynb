```python
import pandas as pd
import matplotlib.pyplot as plt
url='https://raw.githubusercontent.com/SCOREnetworkorg/sports-data-repository/refs/heads/main/data/nba_draft.csv'
df=pd.read_csv(url)
#print(df)
df.head()
mean_val=df['total_pts'].mean()
print(f"mean value ={mean_val}")
median_val=df['total_pts'].median()
print(f"meadian value ={median_val}")
mode_val=df['total_pts'].mode()
print(f"mode value ={mode_val}")
var_val=df['total_pts'].var()
print(f"variance value ={var_val}")
sd_val=df['total_pts'].std()
print(f"standard deviation value ={sd_val}")
df['total_pts'].value_counts().plot(kind='kde', title='Total point count')
plt.xlabel('total_pts')
plt.ylabel('count')
plt.show()
df['total_pts'].value_counts().plot(kind='hist', title='Total point count')
plt.xlabel('total_pts')
plt.ylabel('count')
plt.show()
```
